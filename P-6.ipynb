## P-6A.Line Plot

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,10,0.1)

data1=np.sin(x)
data2=np.cos(x)

plt.plot(data1)
plt.plot(data2)


plt.xlabel('x label')
plt.ylabel('y label')
plt.legend(['Marker1','Marker2'])
plt.title('Chart Tile')
plt.show()
```

##  P-6.B)Bar Plot

```python
import numpy as np
import matplotlib.pyplot as plt

counts = [979, 120, 12]
fuelType = ('Petrol', 'Diesel', 'CNG')
index = np.arange(len(fuelType))

plt.bar(index, counts, color=['red', 'blue', 'cyan'])##

plt.xticks(index, fuelType, rotation='vertical')##

plt.title('Bar plot of fuel types')
plt.xlabel('Fuel Types')
plt.ylabel('Frequency')
plt.legend(['Petrol', 'Diesel', 'CNG'])
plt.show()
```

##  P-6.D)Histogram Plot

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

#plt.hist(data) #basic
plt.hist(data, bins=30, color='skyblue', edgecolor='black') #extra

# Adding Labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')

# Display the plot
plt.show()
```
